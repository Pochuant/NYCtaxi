```python
import google.datalab.bigquery as bq
import pandas as pd
```

```python
%%bq query
#standardSQL
SELECT
  TIMESTAMP_TRUNC(pickup_datetime,
    DAY) day,
  SUM(passenger_count) AS TotalPassengersNum
FROM
  `bigquery-public-data.new_york.tlc_yellow_trips_20*`
WHERE _TABLE_SUFFIX BETWEEN '09' AND '14'
GROUP BY
  day
ORDER BY 
  day ASC
```

```python
totalPassengerSingleDay = bq.Query('SELECT TIMESTAMP_TRUNC(pickup_datetime, DAY) day, SUM(passenger_count) AS TotalPassengersNum FROM `bigquery-public-data.new_york.tlc_yellow_trips_20*` WHERE _TABLE_SUFFIX BETWEEN "09" AND "14" GROUP BY day ORDER BY day ASC')
df = totalPassengerSingleDay.execute(output_options=bq.QueryOutput.dataframe()).result()
df.head(10)
```
